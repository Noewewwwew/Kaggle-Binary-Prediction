```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import xgboost as xgb
from xgboost import plot_importance

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('/content/train.csv')
data = pd.read_csv('/content/test.csv')

df = df.drop('id', axis=1)
data = data.drop('id', axis=1)
print(df['class'].value_counts())
df.head(3)
```

```python
from sklearn.preprocessing import LabelEncoder
import numpy as np

class SafeLabelEncoder(LabelEncoder):
    def __init__(self):
        super().__init__()
        self.classes_ = np.array([])

    def fit(self, y):
        y = y.astype(str)
        self.classes_ = np.unique(y)
        return self

    def transform(self, y):
        y = y.astype(str)
        # 새롭게 등장한 카테고리(클래스)에 대해 -1을 반환
        return np.array([self._safe_transform(val) for val in y])

    def _safe_transform(self, val):
        if val in self.classes_:
            return np.where(self.classes_ == val)[0][0]
        else:
            # 새로운 데이터에 대해 -1을 반환
            return -1

    def fit_transform(self, y):
        return self.fit(y).transform(y)
```

```python
Pass = set(['cap-diameter' , 'stem-height' , 'stem-width' , 'class'])
le = SafeLabelEncoder()
for key in df.columns:
    print(key)
    if key in Pass : continue
    le.fit(df[key])
    df[key] = le.transform(df[key])
    data[key] = le.transform(data[key])
df['class'] = le.fit_transform(df['class'])
y = df['class']
X = df.drop('class', axis=1)
print(y.value_counts())
```

```python
# X_train , X_test , y_train , y_test = train_test_split(X , y , test_size=0.1 , random_state=156)
X_train , X_eval , y_train , y_eval = train_test_split(X , y , test_size=0.1 , random_state=156)
```

```python
from hyperopt import hp

bayesian_params = {'max_depth' : hp.quniform('max_depth' , 6 , 20 , 1) ,
                    'num_leaves' : hp.quniform('num_leaves' , 24 , 64 , 1) ,
                    'max_bin' : hp.quniform('max_bin' , 10 , 500 , 1) ,
                    'min_child_weight' : hp.quniform('min_child_weight' , 1 , 100 , 1) ,
                    'learning_rate' : hp.uniform('learning_rate' , 0.01 , 0.3) ,
                    'reg_lambda' : hp.uniform('reg_lambda' , 0.001 , 10) ,
                    'reg_alpha' : hp.uniform('reg_alpha' , 0.01 , 50) ,
                    'subsample' : hp.uniform('subsample' , 0.5 , 1) ,
                    'colsample_bytree' : hp.uniform('colsample_bytree' , 0.5 , 1)
                    }
```

```python
from lightgbm import LGBMClassifier
from hyperopt import STATUS_OK
from sklearn.model_selection import cross_val_score

def objective_func(search_space):
    lgbm = LGBMClassifier(
        n_estimators = 100,
        max_depth = int(search_space['max_depth']),
        num_leaves = int(search_space['num_leaves']),
        max_bin = int(search_space['max_bin']),
        min_child_weight = int(search_space['min_child_weight']),
        learning_rate = search_space['learning_rate'],
        reg_lambda = search_space['reg_lambda'],
        reg_alpha = search_space['reg_alpha'],
        subsample = search_space['subsample'],
        colsample_bytree = search_space['colsample_bytree'],
        eval_metric = 'logloss',
        tree_method='gpu_hist',  # GPU 사용 설정
        use_label_encoder=False,  # XGBoost 경고 방지를 위한 설정
        n_jobs=-1  # 모든 CPU 코어 사용
    )

    accuracy = cross_val_score(lgbm , X_train , y_train , scoring='accuracy' , cv = 3)

    return {'loss' : -1 * np.mean(accuracy) , 'status' : STATUS_OK}


```

```python
from hyperopt import fmin , tpe , Trials

trial_val = Trials()

best = fmin(
    fn = objective_func,
    space = bayesian_params,
    algo = tpe.suggest,
    max_evals = 50,
    trials = trial_val
)

best
```

```python
lgb_wrapper = LGBMClassifier(n_estimators = 500 ,
                            learning_rate = round(best['learning_rate'] , 5),
                            max_depth = int(best['max_depth']),
                            min_child_weight = int(best['min_child_weight']),
                            colsample_bytree = round(best['colsample_bytree'] , 5),
                            reg_lambda = round(best['reg_lambda'] , 5),
                            reg_alpha = round(best['reg_alpha'] , 5),
                            subsample = round(best['subsample'] , 5),
                            num_leaves = int(best['num_leaves']),
                            max_bin = int(best['max_bin']),
                            early_stopping_rounds=50,
                            eval_metric='logloss'
                            )

evals = [(X_train,y_train) , (X_eval , y_eval)]
lgb_wrapper.fit(X_train , y_train  , eval_set=evals)
```

```python
pred_probs_lgbm = lgb_wrapper.predict(data)
# preds = [1 if x > 0.5 else 0 for x in pred_probs]

# preds[:5]
```

```python
from xgboost import XGBClassifier
xgb_wrapper = XGBClassifier(n_estimators = 500 ,
                            learning_rate = round(0.08094810900900441 , 5),
                            max_depth = int(19),
                            min_child_weight = int(2),
                            colsample_bytree = round(0.5395317035035201 , 5),
                            early_stopping_rounds=50,
                            eval_metric='logloss'
                            )


evals = [(X_train,y_train) , (X_eval , y_eval)]
xgb_wrapper.fit(X_train , y_train  , eval_set=evals , verbose=True)
```

```python
pred_probs_xgb = xgb_wrapper.predict(data)
```

```python
preds = [1 if (lg * 0.2 + xg * 0.8) > 0.5 else 0 for lg ,xg in zip(pred_probs_lgbm , pred_probs_xgb)]
```
