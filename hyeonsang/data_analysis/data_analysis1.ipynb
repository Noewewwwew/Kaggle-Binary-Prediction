veil-type:

버섯의 발달 과정 중에 갓을 덮고 있는 막(veil)의 유형을 말합니다. 주로 두 가지가 있으며, 부분막(partial veil)은 갓과 줄기 사이를 덮고, 전면막(universal veil)은 갓과 줄기를 모두 덮습니다.

spore-print-color:

버섯의 포자가 떨어져 만든 인쇄물의 색을 말합니다. 이 색상은 버섯을 식별하는 중요한 특징입니다.

stem-root:

버섯의 줄기가 땅에 어떻게 뿌리를 내리고 있는지를 설명합니다. 뿌리가 깊거나 얕을 수 있습니다.

veil-color:

버섯의 막(veil)의 색상입니다. 이 색상은 버섯의 다른 부분과 다를 수 있으며, 식별에 도움이 됩니다.

stem-surface:

버섯 줄기의 표면 질감과 외관을 말합니다. 예를 들어, 매끄럽거나, 비늘이 있거나, 섬유질일 수 있습니다.

gill-spacing:

버섯의 갓 아래에 있는 아가미(gills) 간의 간격을 말합니다. 아가미는 조밀하게 붙어 있거나 드물게 붙어 있을 수 있습니다.

cap-surface:

버섯의 갓 표면의 질감과 외관을 말합니다. 예를 들어, 매끄럽거나, 비늘이 있거나, 점액질일 수 있습니다.

gill-attachment:

버섯의 아가미가 줄기에 어떻게 붙어 있는지를 설명합니다. 예를 들어, 아가미가 줄기에 붙어 있지 않거나, 살짝 붙어 있거나, 줄기를 따라 내려갈 수 있습니다.

ring-type:

버섯의 줄기에 있는 고리(ring)의 유형을 말합니다. 고리는 부분막이 남은 흔적입니다.

gill-color:

버섯의 아가미 색상입니다. 이 색상은 버섯의 포자 색상과 관련이 있을 수 있습니다.

habitat:

버섯이 자라는 환경을 말합니다. 예를 들어, 숲, 목초지, 죽은 나무 등에서 자랄 수 있습니다.

cap-shape:

버섯의 갓 모양을 말합니다. 예를 들어, 평평하거나, 볼록하거나, 오목할 수 있습니다.

stem-color:

버섯 줄기의 색상입니다. 줄기의 색상은 버섯 식별에 중요한 특징입니다.

has-ring:

버섯 줄기에 고리가 있는지 여부를 나타냅니다.

cap-color:

버섯의 갓 색상입니다. 색상은 버섯을 식별하는 데 중요한 요소입니다.

does-bruise-or-bleed:

버섯이 손상되었을 때 변색하거나 액체가 나오는지 여부를 나타냅니다.

cap-diameter:

버섯 갓의 직경입니다. 갓의 크기는 버섯 식별에 유용한 정보입니다.

id:

버섯의 고유 식별 번호입니다. 주로 데이터베이스에서 사용됩니다.

stem-width:

버섯 줄기의 너비입니다. 줄기의 굵기 또한 식별에 중요한 정보입니다.

stem-height:

버섯 줄기의 높이입니다. 줄기의 길이는 버섯의 크기를 결정하는 요소 중 하나입니다.

season:

버섯이 자라는 계절입니다. 특정 버섯은 특정 계절에만 나타날 수 있습니다.
